# Proyecto 3 — Análisis de abandono de clientes en telecomunicaciones

## Carga y exploración del dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv("churn-bigml-80.csv")
df.head()
```

```python
df.info()
```

```python
df["Churn"].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Churn", data=df)

plt.title("Distribución de clientes que abandonan el servicio")
plt.xlabel("Abandono del cliente (Churn)")
plt.ylabel("Número de clientes")

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(x="Churn", y="Customer service calls", data=df)

plt.title("Relación entre abandono de clientes y llamadas al servicio al cliente")
plt.xlabel("Churn")
plt.ylabel("Número de llamadas al servicio al cliente")

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(x="Churn", y="Total day minutes", data=df)

plt.title("Relación entre uso del servicio y abandono del cliente")
plt.xlabel("Churn")
plt.ylabel("Minutos de llamadas durante el día")

plt.show()
```
